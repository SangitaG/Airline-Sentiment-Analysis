### Now lets take a look at different n-gram features, for the different processed text and compare.

### Top Uni-grams - Data type: cleaned text (urls, @users, punctuations...) + stopwords
NOTE: I ran all data types: `clean text only`, `stopwords added` and `stemming added`. Out of all 3 data types, clean text + stopwords seemed to have the most informative one. So this is the result I will present.

````python
# vectorize text with 'only cleaning' applied to it.
cv = CountVectorizer(ngram_range=(1,1))                     
X_cv = cv.fit_transform(proc_df['lemma_stopw_clean_text'])

words_Uni_df = pd.DataFrame(X_cv.toarray(), columns=cv.get_feature_names())

# vectorize text with 'stopwords' applied to it.
cv_stop = CountVectorizer(ngram_range=(2,2))
X_cv_stop = cv_stop.fit_transform(proc_df['lemma_stopw_clean_text'])

words_Bi_df = pd.DataFrame(X_cv_stop.toarray(), columns=cv_stop.get_feature_names())

# vectorize text 'with stemming' applied to it.
cv_stem = CountVectorizer(ngram_range=(3,3))
X_cv_stem = cv_stem.fit_transform(proc_df['lemma_stopw_clean_text'])

words_Tri_df = pd.DataFrame(X_cv_stem.toarray(), columns=cv_stem.get_feature_names())
````

````python
words_sorted_ordimp = words_Uni_df.sum().sort_values(ascending=False)
                                                    
words_sorted_ordimp.head(20).plot(kind='barh', title='Uni-Grams: cleaned text+stopwords+lemmatization', 
                                      figsize=(6,8), fontsize=12)
plt.tight_layout()
plt.savefig(img_dir+'UniGrams.png')
````

````python
words_sorted_ordimp = words_Bi_df.sum().sort_values(ascending=False)
                                                    
words_sorted_ordimp.head(20).plot(kind='barh', title='Bi-Grams: cleaned text+stopwords+lemmatization', 
                                      figsize=(6,8), fontsize=12)
plt.tight_layout()
plt.savefig(img_dir+'BiGrams.png')
````

````python
words_sorted_ordimp = words_Tri_df.sum().sort_values(ascending=False)
                                                    
words_sorted_ordimp.head(20).plot(kind='barh', title='Tri-Grams: cleaned text+stopwords+lemmatization', 
                                      figsize=(6,8), fontsize=12)
plt.tight_layout()
plt.savefig(img_dir+'TriGrams.png')
````
