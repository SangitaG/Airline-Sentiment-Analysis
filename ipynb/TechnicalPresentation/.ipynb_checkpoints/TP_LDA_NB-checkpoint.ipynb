### What is LDA?

Latent Dirichlet allocation (LDA) is a topic model that generates topics based on word frequency from a set of documents.<br>
LDA is particularly useful for finding reasonably accurate mixtures of topics within a given document set.


### Goal: 
`Generate topics from bodies of text, to help airlines gather information about what  topics people are talking about,`<br>
`regarding their product, service or brand.`

LDA Topic Modelling:<br>
1. Preprocessing of text.
  1. clean text (urls, punctuations, hastags...)
  2. Remove punctuation.
  3. Remove stopwords.
  4. Lemmatize text.
2. Create an LDA model.
