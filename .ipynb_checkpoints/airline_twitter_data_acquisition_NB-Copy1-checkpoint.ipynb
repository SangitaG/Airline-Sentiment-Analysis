```python
# The following method collects tweets for a give airline and stores in a csvfile.
```

```python
def pull_tweets_store_csv(airline, filename):
    
    # Open/Create a file to append data
    csvFile = open(filename + '.csv', 'a')
    #Use csv Writer
    csvWriter = csv.writer(csvFile)

    for tweet in tweepy.Cursor(api.search,q="#"+airline,count=1000,
                               lang="en",
                               tweet_mode='extended',
                               since="2017-11-26").items():
        
#        print (tweet.created_at, tweet.full_text)
        csvWriter.writerow([airline, tweet.created_at, tweet.full_text.encode('utf-8')])
        
```

```python
# Airline names: UnitedAirlines, AmericanAirlines, SouthWestAirlines, 
# VirginAmerica, USAirways, JetBlue
```

```python
pull_tweets_store_csv("UnitedAirlines", "test")
pull_tweets_store_csv("AmericanAirlines", "test")
pull_tweets_store_csv("SouthWestAirlines", "test")
pull_tweets_store_csv("VirginAmerica", "test")
pull_tweets_store_csv("USAirways", "test")
pull_tweets_store_csv("JetBlue", "test")
```
